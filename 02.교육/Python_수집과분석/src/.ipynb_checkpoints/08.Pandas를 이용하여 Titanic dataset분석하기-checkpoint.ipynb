### 컬럼을 추가 수정  삭제

```python
#컬럼을 추가하고 초기화한다.
train['DataCategory']='Titanic'
```

```python
train.head(5)
```

```python
train['Id2'] = range(0, 891)
```

```python
train.tail(3)
```

```python
#기존 컬럼을 가져와소ㅓ 새로은 컬럼 만들기
train['FamilySize']=train['SibSp']+train['Parch'] + 1
```

```python
train[['FamilySize','SibSp','Parch']].head(5)
```

```python
#bool타립의 컬럼을 추가
train['Natitionity_FR'] = train['Embarked'] == 'C'
train['Natitionity_UK'] = train['Embarked'].isin(['S','Q'])
train[['Embarked','Natitionity_FR', 'Natitionity_UK']].head(5)
```

```python
#국적이 프랑스만
train.loc[train['Natitionity_FR']].head(5)
```

```python
train.loc[train['Natitionity_FR'], "Natitionity"] = 'France'
```

```python
train.head(3)
```

```python
train.loc[train['Natitionity_UK'], "Natitionity"] = 'Eagland'
```

```python
train.head(3)
```

```python
#Age의 평균값을 계산해서 null을 채워 넣는다
mean_age = train['Age'].mean()
train.loc[train['Age'].isnull(), 'Age'] = mean_age
```

```python
train.info()
```

```python
#1개의 컬럼 삭제
del train['Id']
```

```python
train.columns
```

```python
#한번에 여러개의 컬럼삭제는 drop사용
train.drop(['DataCategory','Natitionity'],axis=1, inplace=True)
```

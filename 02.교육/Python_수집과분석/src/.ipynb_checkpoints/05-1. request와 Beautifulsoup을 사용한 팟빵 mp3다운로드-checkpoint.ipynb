```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from itertools import count
from time import sleep


def get_list(pid):
    #page = 1
    
    #1씩증가하면서 처리.....반복..1이면 계속처리
    for page in count(1):
        
        params = {
            'id':pid,
            'page':page
        }
        page_url='http://www.podbbang.com/podbbangchnew/episode_list'

        #이 싸이트는 이게 있어야 함
        headers = {
            'Referer':'http://www.podbbang.com/ch/{pid}'.format(pid=pid)
        }
        response = requests.get(page_url,params=params)
        response.encoding = 'utf-8'
        #response.encoding

        #text는 text 바이너리는 content
        html = response.text

        soup = BeautifulSoup(html, 'html.parser')

        for li_tag in soup('li'):
            try:        
                #에러처리할 코드
                title = li_tag.find('dt')['title']  
                link = urljoin(page_url, li_tag.find('a')['href'])
                print(title, link)
            except(TypeError, KeyError):
                #에러날때 실행
                print("Error 발생")
                return None
            else:     
                #정상일때 실행
                print('download link : {}'.format(link))
                '''
                #mp3파일을 저장한다.
                mp3_bin = requests.get(link, headers=headers ).content
                filenmae = '{}.mp3'.format(title);
                with open(title,'wb') as f:
                    print('image file name {} ( {} bytes) ', format(file_name, len(mp3_bin)) )
                    f.write(mp3_bin)
                break
                '''
            sleep(0.5)
    
```

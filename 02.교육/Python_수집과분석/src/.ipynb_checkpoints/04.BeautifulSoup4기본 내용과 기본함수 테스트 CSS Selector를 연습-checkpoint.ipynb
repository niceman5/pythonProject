### 멜론 100 chart title, link 출력하기

```python
import requests
from bs4 import BeautifulSoup
import re


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36'
}
url = "http://www.melon.com/chart/index.htm"

#heaer에 'User-Agent'가 없으면 응답이 안옴......싸이트 마다 자동처리를 막기 위해
html = requests.get(url, 'html.parser', headers=headers).text
soup = BeautifulSoup(html, 'html.parser')
song_list = soup.select('#tb_list tr .wrap_song_info a[href*=playSong]')
for idx, song in enumerate(song_list):
    title = song.text
    link = song['href']
    matched = re.search(r"(\d+)\);", link)
    #print(matched)
    if matched:
        #print(matched.group(0), matched.group(1))
        songid = matched.group(1)
        songUrl = 'http://www.melon.com/song/detail.htm?songId=' + songid
    print(idx+1, title, songUrl)
```

### 국회의원 이름과 번호 출력하기

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'http://www.assembly.go.kr/assm/memact/congressman/memCond/memCondListAjax.do?currentPage=1&rowPerPage=500'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
membetrs = soup.select('.memberna_list dl dt a') #class라서 .으로 시작
for idx, member in enumerate(membetrs, 1):
    #print(member)
    name = member.text
    link = member['href']
    matched = re.search(r"\d+", link)
    if matched:
        member_id = matched.group(0)
        print(idx, name, member_id)
    
        
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'http://www.assembly.go.kr/assm/memact/congressman/memCond/memCondListAjax.do?currentPage=1&rowPerPage=500'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
membetrs = soup.select('.memberna_list dl') #class라서 .으로 시작
for idx, member in enumerate(membetrs, 1):
    #print(idx, member)
    member_info = member.select('dt a')
    member_regin = member.select('.ht')
    
    #print(member_info)
    #print(member_regin[0].text)
    
    name = member_info[0].text
    link = member_info[0]['href']
    matched = re.search(r"\d+", link)
    if matched:
        member_id = matched.group(0)
        print(idx, name, member_id, member_regin[0].text)
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'http://www.assembly.go.kr/assm/memact/congressman/memCond/memCondListAjax.do?currentPage=1&rowPerPage=500'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
membetrs = soup.select('.memberna_list dl') #class라서 .으로 시작
for idx, member in enumerate(membetrs, 1):
    #print(idx, member)
    a_tag = member.select('dt a')
   
    name = a_tag[0].text
    link = a_tag[0]['href']
    matched = re.search(r"\d+", link)
    
    if matched:
        member_id = matched.group(0)
    else:
        member_id = None
    
    dd_tag = member.select('dd.ht')
    area_name = dd_tag[0].text
    print(idx, name, member_id, area_name)
```
